```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
##load the data
data = pd.read_csv('data/train.csv')
data
```

```python
data.columns.to_list()
```

```python
# Dataset map
from pathlib import Path

DATA_DIR = Path('data')
train = data

files = {
    'train': DATA_DIR / 'train.csv',
    'test': DATA_DIR / 'test.csv',
    'supplemental': DATA_DIR / 'supplemental_data.csv',
    'sample_submission': DATA_DIR / 'sample_submission.csv',
    'hospital_ref': DATA_DIR / 'hospital_reference.csv',
    'icd10_ref': DATA_DIR / 'icd10_reference.csv',
}

file_summary = []
for name, path in files.items():
    header = pd.read_csv(path, nrows=0)
    file_summary.append({
        'file': name,
        'size_mb': round(path.stat().st_size / 1024**2, 2),
        'columns': len(header.columns),
    })

pd.DataFrame(file_summary)
```

```python
# Train/test schema and target columns
target_cols = ['adverse_outcome', 'readmission_30d']
post_outcome_cols = ['disposition']  # present in train, absent in test; do not use as a model feature
id_cols = ['encounter_id']

test_cols = pd.read_csv(DATA_DIR / 'test.csv', nrows=0).columns.tolist()
train_only = sorted(set(train.columns) - set(test_cols))
test_only = sorted(set(test_cols) - set(train.columns))

print('train shape:', train.shape)
print('train-only columns:', train_only)
print('test-only columns:', test_only)
print('sample submission columns:', pd.read_csv(DATA_DIR / 'sample_submission.csv', nrows=0).columns.tolist())
```

```python
# Missing values and target balance
missing = train.isna().mean().mul(100).sort_values(ascending=False)
display(missing[missing > 0].to_frame('missing_pct'))

target_balance = pd.concat(
    [train[col].value_counts(normalize=True).mul(100).rename(col) for col in target_cols],
    axis=1,
).round(2)
display(target_balance)

fig, axes = plt.subplots(1, 2, figsize=(9, 3.5))
for ax, col in zip(axes, target_cols):
    sns.countplot(data=train, x=col, ax=ax)
    ax.set_title(col)
    ax.set_xlabel('')
    ax.set_ylabel('count')
plt.tight_layout()
```

```python
# Target rates by clinically meaningful groups
def target_rates(df, group_col, min_count=500):
    counts = df[group_col].value_counts().rename('n')
    rates = df.groupby(group_col)[target_cols].mean().mul(100)
    out = counts.to_frame().join(rates).sort_values('n', ascending=False)
    return out[out['n'] >= min_count].round(2)

for col in ['triage_level', 'age_group', 'mews_score', 'news2_score', 'chief_complaint', 'primary_icd10_code']:
    print('\n', col)
    display(target_rates(train, col).head(15))
```

```python
# Rough feature groups for modeling
feature_cols = [c for c in train.columns if c not in target_cols + post_outcome_cols + id_cols]
groups = {
    'time_location_hospital': [c for c in feature_cols if c.startswith('arrival_') or c in ['city', 'nuts2_region', 'latitude', 'longitude', 'urban_rural', 'hospital_type', 'annual_ed_volume', 'bed_count_category']],
    'demographics': [c for c in feature_cols if c in ['age', 'age_group', 'gender', 'is_pediatric', 'is_geriatric', 'height_cm', 'weight_kg', 'bmi', 'insurance_type', 'ses_index']],
    'triage_presentation': [c for c in feature_cols if c in ['arrival_mode', 'triage_level', 'triage_label', 'chief_complaint']],
    'vitals_labs_scores': [c for c in feature_cols if any(token in c for token in ['bpm', 'mmhg', 'rpm', 'celsius', 'spo2', 'gcs', 'score', 'mgdl', 'meql', 'ul', 'gdl', 'pct', 'inr', 'aptt', 'lactate', 'troponin', 'bnp', 'crp', 'd_dimer', 'albumin', 'bilirubin', 'egfr'])],
    'comorbidity_med_proc': [c for c in feature_cols if c.startswith(('comorbidity_', 'med_', 'proc_')) or c in ['polypharmacy_5plus', 'iv_fluids_ml', 'consult_requested', 'consult_specialty']],
    'covid_diagnosis': [c for c in feature_cols if c.startswith('covid_') or c.startswith('primary_') or c == 'pandemic_period'],
}

pd.DataFrame({'group': groups.keys(), 'n_cols': [len(v) for v in groups.values()]})
```

## Modular baseline pipeline

The reusable functions live in `emergency_ml_pipeline.py`. The cells below keep the notebook readable while making the modeling workflow easy to rerun.

```python
from importlib import reload
import emergency_ml_pipeline as em

reload(em)

DATA_DIR = 'data'
RANDOM_STATE = 42
TRAIN_END_YEAR = 2022
INCLUDE_SUPPLEMENTAL = False
DEV_SAMPLE_N = None  # set to 150_000 for a quick first run on a smaller sample

pd.set_option('display.max_columns', 200)
pd.set_option('display.width', 160)

try:
    em._require_sklearn()
    SKLEARN_AVAILABLE = True
    print('scikit-learn is available for modeling.')
except ModuleNotFoundError as err:
    SKLEARN_AVAILABLE = False
    print(err)
    print('EDA cells can still run. For modeling, run: %pip install scikit-learn')
```

```python
# Data loading and schema checks
if 'train' not in globals():
    train = em.load_training_data(
        DATA_DIR,
        include_supplemental=INCLUDE_SUPPLEMENTAL,
        sample_n=DEV_SAMPLE_N,
        random_state=RANDOM_STATE,
    )
elif DEV_SAMPLE_N is not None and DEV_SAMPLE_N < len(train):
    train = train.sample(n=DEV_SAMPLE_N, random_state=RANDOM_STATE).sort_index()

test = em.load_test_data(DATA_DIR)

display(em.summarize_files(DATA_DIR))
schema = em.schema_report(train, test)
print('Train shape:', train.shape)
print('Test shape:', test.shape)
print('Train-only columns:', schema['train_only'])
print('Test-only columns:', schema['test_only'])
```

```python
# Basic EDA: target balance, missingness, and high-level target rates
display(em.target_balance(train))
display(em.missing_summary(train, min_missing_pct=0))

for group_col in ['arrival_year', 'triage_level', 'age_group', 'mews_score', 'news2_score']:
    print(f'\nTarget rates by {group_col}')
    display(em.target_rates(train, group_col).head(20))
```

```python
# Leakage removal, missing indicators, and feature type separation preview
preview_X, preview_y = em.make_feature_target(train, target='adverse_outcome')
numeric_features, categorical_features = em.separate_feature_types(preview_X)

print('Feature matrix shape:', preview_X.shape)
print('Target positive rate:', round(preview_y.mean(), 4))
print('Numeric features:', len(numeric_features))
print('Categorical features:', len(categorical_features))
print('Leakage columns present after prep:', sorted(set(em.LEAKAGE_COLS) & set(preview_X.columns)))
print('Target columns present after prep:', sorted(set(em.TARGET_COLS) & set(preview_X.columns)))

missing_indicator_cols = [c for c in preview_X.columns if c.endswith('_missing')]
display(preview_X[missing_indicator_cols].mean().mul(100).sort_values(ascending=False).to_frame('missing_indicator_rate_pct'))
```

```python
# Time-based train-validation split
train_part, valid_part = em.split_time_based(train, train_end_year=TRAIN_END_YEAR)

split_summary = pd.DataFrame({
    'split': ['train', 'valid'],
    'rows': [len(train_part), len(valid_part)],
    'min_year': [train_part['arrival_year'].min(), valid_part['arrival_year'].min()],
    'max_year': [train_part['arrival_year'].max(), valid_part['arrival_year'].max()],
    'adverse_rate': [train_part['adverse_outcome'].mean(), valid_part['adverse_outcome'].mean()],
    'readmission_rate': [train_part['readmission_30d'].mean(), valid_part['readmission_30d'].mean()],
})

display(split_summary)
```

```python
# Baseline model training for both binary targets
# This uses a sparse one-hot encoded + scaled linear classifier baseline.
if not SKLEARN_AVAILABLE:
    raise RuntimeError('Install scikit-learn first in this notebook kernel: %pip install scikit-learn')

results = em.run_all_baselines(
    train,
    train_end_year=TRAIN_END_YEAR,
    random_state=RANDOM_STATE,
    min_category_frequency=50,
)

display(em.metrics_table(results).round(4))
```

```python
# Evaluation details and feature importance
for target, result in results.items():
    print(f'\n=== {target} ===')
    print('Train feature shape:', result.train_shape)
    print('Valid feature shape:', result.valid_shape)
    print('Numeric/categorical:', len(result.numeric_features), '/', len(result.categorical_features))
    display(result.confusion_matrix)
    display(result.feature_importance)
```

## Submission generation

After validating the baseline, retrain the two target-specific models on the full training data and write predictions in `sample_submission.csv` format.

```python
# Fit final models on all available training rows.
# Keep INCLUDE_SUPPLEMENTAL=False for the first submission unless you decide to include supplemental data.
if not SKLEARN_AVAILABLE:
    raise RuntimeError('Install scikit-learn first in this notebook kernel: %pip install scikit-learn')

final_train = train

final_models = em.fit_full_models(
    final_train,
    random_state=RANDOM_STATE,
    min_category_frequency=50,
)

list(final_models.keys())
```

```python
# Create Kaggle submission using sample_submission.csv column order and encounter_id alignment.
# threshold=None writes probabilities. Use threshold=0.5 only if the competition explicitly wants hard labels.
submission = em.make_submission(
    final_models,
    test,
    sample_submission_path='data/sample_submission.csv',
    threshold=None,
)

output_path = em.save_submission(submission, 'submission_baseline_sgd.csv')
print('Saved:', output_path)
display(submission.head())
display(submission[em.TARGET_COLS].describe())
```
